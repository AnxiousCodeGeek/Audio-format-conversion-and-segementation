## Audio Segemntation

```python
import soundfile as sf
import random
import os

def break_audio(audio_path, output_dir, min_segment_duration=2, max_segment_duration=5, num_segments=60):
    # Create the output directory if it doesn't exist
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)
    
    # Read the audio file
    y, sr = sf.read(audio_path)
    
    # Get duration of the audio file in seconds
    audio_duration = len(y) / sr
    
    for i in range(num_segments):
        # Generate random start time within the audio duration
        start_time = random.uniform(0, audio_duration - max_segment_duration)
        
        # Generate random duration for the segment
        segment_duration = random.uniform(min_segment_duration, max_segment_duration)
        
        # Extract segment from the audio
        start_frame = int(start_time * sr)
        end_frame = int((start_time + segment_duration) * sr)
        segment = y[start_frame:end_frame]
        
        # Save the segment to a WAV file
        segment_output_path = os.path.join(output_dir, f"segment_{i}.wav")
        sf.write(segment_output_path, segment, sr)
        
        print(f"Segment {i+1}: Start Time = {start_time:.2f}, Duration = {segment_duration:.2f} seconds")
```
